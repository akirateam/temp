```python
import torch
from transformers import pipeline
import os
```

```python
dolly_pipeline = pipeline(model="databricks/dolly-v2-3b",
                          torch_dtype=torch.bfloat16,
                          trust_remote_code=True,
                          device_map="auto")
```

```python
def get_completion_dolly(input):
  system = f"""
  Faça sua pergunga sobre física ao AI Clau Clau utizando frases simples.
  Ele o ajudará na media do possível, daquele jeitão dele.
  """
  prompt = f"#### System: {system}\n#### User: \n{input}\n\n#### Resposta do AI Clau Clau:"
  print(prompt)
  dolly_response = dolly_pipeline(prompt, max_new_tokens = 500)
  return dolly_response[0]["generated_text"]
```

```python
#Explain to me the difference between nuclear fission and fusion.
prompt = "Me forneça mais detalhes sobre o princípio Bluemercury"
print(get_completion_dolly(prompt))
```
